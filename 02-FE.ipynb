```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('cleaned.csv')
data.head()
```

```python
def convert_price(x):
    x = str(x).lower().strip()

    if 'cr' in x or 'crore' in x:
        num = float(x.split()[0])
        return num * 10000000   

    elif 'l' in x or 'lakh' in x:
        num = float(x.split()[0])
        return num * 100000     

    else:
        return None

data['price_rupees'] = data['Price'].apply(convert_price)

data['price_per_sqft'] = data['price_rupees'] / data['Area in Sqrt']
```

```python
data['price_rupees']=data['price_rupees'].astype(int)
```

```python
data.head()
```

```python
data = data.drop(columns='Price')
```

```python
data['area_per_bhk'] = data['Area in Sqrt'] / data['No.of Bhk']
```

```python
data.to_csv('cleaned1.csv',index=False)
```
