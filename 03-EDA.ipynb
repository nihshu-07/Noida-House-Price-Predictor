```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('cleaned1.csv')
data.head()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['area_per_bhk'], bins=30)
plt.xlabel('Area Per BHK')
plt.ylabel('No. of Houses')
plt.title('Distribution of Area per BHK')
plt.show()
```

```python
plt.figure(figsize=(8,2))
plt.boxplot(data['area_per_bhk'], vert=False)
plt.xlabel('Area per BHK (sq ft)')
plt.title('Boxplot of Area per BHK')
plt.show()
```

```python
data['Location'].value_counts().plot(kind='bar', figsize=(12,5))
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Number of Houses by Location')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(data['Area in Sqrt'], data['price_rupees'], alpha=0.5)
plt.xlabel('Area in Sq Ft')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['price_per_sqft'], bins=30)
plt.xlabel('Price per Sq Ft')
plt.ylabel('Count')
plt.title('Distribution of Price per Sq Ft')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data['price_rupees'], bins=30)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Distribution of House Prices')
plt.show()
```

```python
plt.figure(figsize=(8,2))
plt.boxplot(data['price_rupees'], vert=False)
plt.xlabel('Price')
plt.title('Price Outliers')
plt.show()
```
